```python
# Bring in Pandas Library
import pandas as pd
```

```python
# Load the Data. And check that its loaded
df = pd.read_excel("F:/Data Analyst/data sets/Customer Call List.xlsx")
df
```

```python
# drop duplicates and save df without dupes

df = df.drop_duplicates()
og_df = df
df
```

```python
# Remove Not_Useful_Column and save df
df = df.drop(columns="Not_Useful_Column")
df
```

```python
# Remove extra characters from last_name
# .str.lstrip()  removes from left side
# .str.rstrip() removes from right side

df["Last_Name"] = df["Last_Name"].str.strip("./_")
df
```

```python
# Remove every character that is NOT a letter or number
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
df
```

```python
# Change each number to a string
# Then apply lamda to add a - between the specific spaces
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] +'-' + x[3:6] +'-' + x[6:10])
df
```

```python
# Remove the NaN and Na values
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','').str.replace('Na--','')
df
```

```python
# Split the address's. Luckily they are separated by a comma

df[["Street_Address","State","Zip_Code"]] = df["Address"].str.split(',',2, expand=True)
df
```

```python
# Replace all Yes and No with Y and N

df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y').str.replace('No','N')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y').str.replace('No','N')

df
```

```python
# Remove N/a Values

df = df.replace('N/a','')
df = df.fillna('')
df
```

```python
# Rename Address column to Full Address

df = df.rename(columns={'Address': 'Full_Address'})
df
```

```python
# Loop through and remove Do_Not_Contact customers

for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
df
```

```python
# Fill out rest of Do_not_Contact

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('','N').str.replace('NNN','N')
df
```

```python
# Loop through and remove customers without a Phone Number

# Could have also used df = df.dropna(subset=["Phone_Number"], inplace=True)

for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)
df
```

```python
df = df.reset_index(drop=True)
df
```
