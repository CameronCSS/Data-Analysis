# Data Visualization using Python

You can get the dataset from [HERE](https://github.com/AlexTheAnalyst/PandasYouTubeSeries/blob/main/Ice%20Cream%20Ratings.csv)

```python
# Import all libraries you might use or need for Visualization

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in your dataset using the appropriate read_ function

df = pd.read_csv(r"F:\Data Analyst\data sets\Alex Pandas\pandas basics\Ice Cream Ratings.csv")
df
```

```python
# Different plot styles

print(plt.style.available)
```

```python
# Using a plot style

plt.style.use('dark_background')
```

```python
# Show basic plot. By default a Line plot is used

df.plot()
```

```python
# Set the Index to the Date

df = df.set_index('Date')
df
```

```python
# Changing the plot style back to default

plt.style.use('default')
```

```python
# Show basic plot. By default a Line plot is used

df.plot()
```

```python
# Add customization to change the default lok

df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
# Bar plot

df.plot(kind = 'bar')
```

```python
# Bar chart

df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
# Bar chart

df.plot(kind = 'bar', stacked = True)
```

```python
# Bar chart

df.plot.barh(stacked = True)
```

```python
# Scatterplot

df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating')
```

```python
# Scatterplot

df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 100)
```

```python
# Scatterplot

df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500)
```

```python
# Scatterplot

df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500, c = 'Green')
```

```python
# Histogram

df.plot.hist()
```

```python
# Histogram

df.plot.hist(bins = 20)
```

```python
# Histogram bins default is 10

df.plot.hist(bins = 20)
```

```python
# Boxplot

df.boxplot()
```

```python
# Area plot
## Figsize can be used to increase the size of the graph if you need more room to show your data

df.plot.area(figsize = (10,5))
```

```python
# The amazing incredible piechart

df.plot.pie(y='Flavor Rating', figsize = (10,6))
```
